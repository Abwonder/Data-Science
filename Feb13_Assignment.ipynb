1. Only Throw Exceptions for Current Level
2. Don’t Overwhelming Exceptions

6. Avoid Raising Generic Exceptions
7. Raise Custom Exceptions -  Plan for contigencies
8. Define Your Own Exception Hierarchy
9. Document All Exceptions Thrown by a Function -- logging
