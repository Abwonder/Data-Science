```python
logging.basicConfig(filename = "test13.log" , level = logging.DEBUG, format = '%(asctime)s %(name)s %(levelname)s  %(message)s')
```

```python
logging.info("this is my info log")
```

```python
l = [1,2,3,4,[4,5,6] , "sudh" ,"kumar"]  
# create two list of all numbers and second all strings from the above list
```

```python
l1_int = []
l2_str = []
for i in l : 
    logging.info("this is the start of my first for loop {}".format(l))
    logging.info("this is the value of i am logging {}".format(i))
    if type(i) == list:
        for j in i :
            logging.info("logggin my j {j} and i is {i}".format(i = i ,j = j))
            if type(j) == int :
                l1_int.append(j)
    elif type(i) == int :
        l1_int.append(i)
        
    else :
        if type(i) == str:
            l2_str.append(i)
logging.info("this is my final result  with all int {l1} ,with all str{l2}".format(l1 =l1_int ,l2 =l2_str ))
        
```

```python
logging.shutdown()
```
